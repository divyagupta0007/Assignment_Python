```python
import pandas as pd

df = pd.read_csv("cardekho.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_duplicates = df.duplicated().sum()
print(df_duplicates)

if (df_duplicates > 0):

    df = df.drop_duplicates().reset_index(drop=True)
```

```python
if df["engine"].isnull().any():

    df = df.dropna(subset=["engine"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y="engine", color="red")
plt.title("Engine")

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y="selling_price", color="green")
plt.title("Selling Price")
plt.show()


plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y="km_driven", color="red")
plt.title("Km Driven")

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y="selling_price", color="green")
plt.title("Selling Price")
plt.show()


plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y="mileage(km/ltr/kg)", color="red")
plt.title("Mileage")

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y="selling_price", color="green")
plt.title("Selling Price")
plt.show()
```

```python
Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]

print(df.shape[0] - df_filtered.shape[0])



print(df_filtered)

plt.figure(figsize=(10, 10))
sns.heatmap(data=df_filtered.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()


plt.figure(figsize=(8,8))
sns.scatterplot(data=df_filtered, x="km_driven", y="selling_price")
plt.xlabel("Km Driven")
plt.ylabel("Selling Price")
plt.title("Km Driven Vs Selling Price")
```

```python
Q1 = df["mileage(km/ltr/kg)"].quantile(0.25)
Q3 = df["mileage(km/ltr/kg)"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df["mileage(km/ltr/kg)"] >= lower_bound) & (df["mileage(km/ltr/kg)"] <= upper_bound)]

print(df.shape[0] - df_filtered.shape[0])

import matplotlib.pyplot as plt
import seaborn as sns

print(df_filtered)
plt.figure(figsize=(10, 10))
sns.heatmap(data=df_filtered.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

plt.figure(figsize=(8,8))
sns.scatterplot(data=df_filtered, x="mileage(km/ltr/kg)", y="selling_price")
plt.xlabel("Mileage")
plt.ylabel("Selling Price")
plt.title("Mileage Vs Selling Price")
```

```python
Q1 = df["engine"].quantile(0.25)
Q3 = df["engine"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df["engine"] >= lower_bound) & (df["engine"] <= upper_bound)]

print(df.shape[0] - df_filtered.shape[0])

import matplotlib.pyplot as plt
import seaborn as sns

print(df_filtered)
plt.figure(figsize=(10, 10))
sns.heatmap(data=df_filtered.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()


plt.figure(figsize=(8,8))
sns.scatterplot(data=df_filtered, x="engine", y="selling_price")
plt.xlabel("Engine")
plt.ylabel("Selling Price")
plt.title("Engine Vs Selling Price")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df_filtered[["engine", "km_driven", "mileage(km/ltr/kg)"]]
y = df_filtered["selling_price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_test,y_test)

y_pred = model.predict(x_test)

print(f"R-squared Score (R²): {r2_score(y_test, y_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
```
